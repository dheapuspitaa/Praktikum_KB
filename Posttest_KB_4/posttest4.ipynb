Import Library

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('income.csv')

data
```

Import Library Train_Test_Split

```python
from sklearn.model_selection import train_test_split
```

Melakukan Standarisasi Data

```python
from sklearn.preprocessing import StandardScaler
```

Memanggil StandardScaler

```python
Scaler = StandardScaler()
```

Melakukan Data Cleaning

Melakukan Data Cleaning Pada Nilai Null

Membuat Copy Dataset Untuk Data Cleaning

```python
dataCopy = data.copy()

dataCopy
```

Cek Nilai Null Di Dalam Dataset

```python
print("Jumlah Record Bernilai Null : ")

print(dataCopy.isna().sum())
```

Membuat Kolom Attribute Memiliki Nilai Null

```python
dataCopy.loc[[4, 112], 'age'] = None
dataCopy.loc[[15, 220], 'fnlwgt'] = None
dataCopy.loc[[5, 110], 'education.num'] = None
```

Menampilkan Data Yang Sudah Diubah Secara Acak Nilai Pada Attributnya Menjadi Null

```python
dataCopy
```

Cek Kembali Nilai Null

```python
dataCopy.isna().sum()
```

Mengganti Nilai Null Dengan Strategy Menggunakan 3 Nilai Netrik Pengukuran

Import Library SimpleImputer

```python
from sklearn.impute import SimpleImputer
```

```python
mean = SimpleImputer(strategy = 'mean')
median = SimpleImputer(strategy = 'median')
modus = SimpleImputer(strategy = 'most_frequent')

dataCopy['age'] = mean.fit_transform(dataCopy[['age']])
dataCopy['fnlwgt'] = median.fit_transform(dataCopy[['fnlwgt']])
dataCopy['education.num'] = modus.fit_transform(dataCopy[['education.num']])
```

Menampilkan Dan Mengecek Kembali Data Setelah Diubah Nilai Nullnya

```python
dataCopy
```

```python
dataCopy.isna().sum()
```

```python
dt = pd.read_csv('income.csv')
```

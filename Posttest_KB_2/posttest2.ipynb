Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka

```python
#Jumlah record
jumlahRecord = len(newsData[['tweet_num', 'real']]) #Menampilkan jumlah record pada kedua attribute

#Nama attribute
attribute = list(newsData.columns[newsData.dtypes == 'int64']) #Membuat list dari kolom attribute bertipe integer
attribute = "\n".join(attribute) 

#Jumlah attribute
jumlahAttribute = len(newsData.dtypes[newsData.dtypes == 'int64']) #Menghitung banyak attribute bertipe integer

#Set dictionary
dictionary = {"Jumlah Record" : [jumlahRecord], "Nama Attribute" : [attribute], "Jumlah Attribute" : [jumlahAttribute]}

#Membuat tabel dengan tabulate
tabel = pd.DataFrame(dictionary)
print(tabulate(tabel, headers='keys', tablefmt='psql', showindex=False))
```

Menampilkan nilai dataset pada baris dimulai dengan 3 digit NIM akhir (085) hingga data terakhir dan kolom ke 4 hingga kolom terakhir menggunakan ke 2 metode slice

```python
#Menampilkan data dari baris record ke 85 hingga akhir dan dari kolom 4 hingga akhir
tampiLoc = newsData.loc[85:, newsData.columns[3]:]  

print("Metode loc() : ")
print(tampiLoc)

#Menampilkan data dari baris record ke 85 hingga akhir dan dari kolom 4 hingga akhir
tampiIloc = newsData.iloc[85:, 3:] 

print("\nMetode iloc() : ")
print(tampiIloc)
```

Menampilkan jumlah label dari semua attribute yang bernilai object

```python
#Mencari attribute bertipe object
attributeObject = newsData.select_dtypes(include=['object']).columns

#Perulangan pada attribute bertipe object dan menampilkan jumlah label dari setiap attribute
for column in attributeObject:
    labels = newsData[column].nunique() #Mengembalikan jumlah nilai unik pada dataset
    print(f"Attribute '{column}' memiliki {labels} label") #Print f untuk mencetak berulang kali sebanyak perulangan terjadi

totalLabel = sum(newsData[column].nunique() for column in attributeObject) #Sum untuk menjumlahkan semua label yang dimiliki attribute
print(f"\nTotal jumlah label di semua attribute bertipe object adalah {totalLabel}.")
```
